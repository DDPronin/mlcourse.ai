```python
import pandas as pd
```

```python
data = pd.read_csv("../../data/adult.data.csv")
data.head()
```

**1. Сколько мужчин и женщин (признак *sex*) представлено в этом наборе данных?**

```python
data["sex"].value_counts()
```

**2. Каков средний возраст (признак *age*) женщин?**

```python
data[data['sex'] == "Female"]['age'].mean()
```

**3. Какова доля граждан Германии (признак *native-country*)?**

```python
data["native-country"].value_counts(normalize=True)["Germany"]
```

**4-5. Каковы средние значения и среднеквадратичные отклонения возраста тех, кто получает более 50K в год (признак *salary*) и тех, кто получает менее 50K в год?**

```python
data[data["salary"] == ">50K"].describe()["age"]
```

```python
data[data["salary"] == "<=50K"].describe()["age"]
```

**6. Правда ли, что люди, которые получают больше 50k, имеют как минимум высшее образование? (признак *education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* или *Doctorate*)**

```python
data['education'].value_counts()
```

```python
highEduList = ['Bachelors', 'Prof-school', 'Assoc-acdm', 'Assoc-voc', 'Masters', 'Doctorate']
highEdu = (data['education'].apply(lambda isHigh: isHigh in highEduList))
if "highEdu" not in data.columns:  # во избежание ошибок, внесем колонку в таблице, только если ее нет
    data.insert(loc=len(data.columns), column='highEdu', value=highEdu)
```

### Да, почти половина тех, кто представлен в выборке и зарабатывает >50K не имеет высшего образования...

```python
data[data['salary'] == '>50K']['highEdu'].value_counts(normalize=True)
```

### Но в выборке тех, у кого есть высшее образование, менее трети...

```python
data['highEdu'].value_counts(normalize=True)
```

### Среди людей с высшем образованием 43% имеют зарплату >50K

```python
data[data['highEdu'] == True]['salary'].value_counts(normalize=True)
```

### А среди тех, кто его не имеют, только 15%

```python
data[data['highEdu'] == False]['salary'].value_counts(normalize=True)
```

### Вывод: высшее образование заметно повышает шанс на достойную оплату труда

**8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак *marital-status*)? Женатыми считаем тех, у кого *marital-status* начинается с *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.**

```python
data['marital-status'].value_counts()
```

```python
data[data['marital-status'].apply(lambda status: status[0:len('Married')] == "Married")]['salary'].value_counts(normalize=True)
```

```python
data[data['marital-status'].apply(lambda status: status[0:len('Married')] != "Married")]['salary'].value_counts(normalize=True)
```
